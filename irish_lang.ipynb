```python
import torch
from transformers import VitsTokenizer, VitsModel, set_seed

tokenizer = VitsTokenizer.from_pretrained("ylacombe/vits_vctk_irish_male")
model = VitsModel.from_pretrained("ylacombe/vits_vctk_irish_male")

inputs = tokenizer(text="dia duit conas atá tú", return_tensors="pt")

set_seed(555)  # make deterministic

with torch.no_grad():
   outputs = model(**inputs)

waveform = outputs.waveform[0]
```

```python
import scipy

scipy.io.wavfile.write("techno.wav", rate=model.config.sampling_rate, data=waveform)
```
